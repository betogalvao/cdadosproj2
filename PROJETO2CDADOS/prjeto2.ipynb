```python
import pandas as pd
import numpy as np
```

```python
def removeSpaces(string):
    string = string.replace(' ','')
    return string

def lowercase(string):
    string = string.lower()
    return string
```

```python
dfBigMacI = pd.read_csv('bigmac.csv')
dfBigMacI['Year'] = pd.DatetimeIndex(dfBigMacI['date']).year

dfBigMac = dfBigMacI[['name', 'Year', 'dollar_price']]
dfBigMac = dfBigMac.loc[dfBigMac['Year'] == 2015]

dfBigMac['name'] = dfBigMac['name'].apply(removeSpaces)
dfBigMac['name'] = dfBigMac['name'].apply(lowercase)

listax = np.arange(805,848,1)

listamedia = []
for i in range(805, 848, 1):
    media = (dfBigMac['dollar_price'][i] + dfBigMac['dollar_price'][i + 43])/2
    listamedia.append(media)

dfBigMac['media'] = listamedia * 2
dfBigMac = dfBigMac.loc[listax,:]
dfBigMac = dfBigMac.drop('dollar_price', axis='columns', inplace=False)
paises = []
for pais in dfBigMac['name']:
    paises.append(pais)
dfBigMac
```

```python
dfObesidadeI = pd.read_csv('obesity-cleaned.csv')
dfObesidade = dfObesidadeI.loc[dfObesidadeI['Year'] == 2015]
dfObesidade = dfObesidade.loc[dfObesidade['Sex'] == 'Both sexes']
dfObesidade.drop('Sex', axis='columns', inplace=True)
dfObesidade.drop('Unnamed: 0', axis='columns', inplace=True)
dfObesidade['Country'] = dfObesidade['Country'].apply(removeSpaces)
dfObesidade['Country'] = dfObesidade['Country'].apply(lowercase)

listaindex = list(dfObesidade.index)

paises1 = []
for pais in dfObesidade['Country']:
    paises1.append(pais)
    
i = 0
for pais in paises:
    if pais in paises1:
        dfObesidade['Country'][int(listaindex[i])] = pais
        i+=1
    elif pais not in paises1:
        dfObesidade['Country'][int(listaindex[i])] = ''
        i+=1
dfObesidade
```
